The situation for in part (a) is fundamentally different from the spherical case since Newton's theorem's don't apply in a 2 dimensional scenario (i.e. thin disk). This can be seen looking the image below which shows how the 2D (rings) and 3D (shells) geometries translates into different relations of the force experienced by an object in the interior. In the spherical case the forces cancel since $M\propto R^2$ and $F\propto 1/R^2$, while in rings $M \propto R$ so forces can't cancel. 

This means that $v_c$ can be written in terms of the enclosed mass, $v_c(R)^2=\frac{GM(<R)}{R}$, for a spherical potential but not for a disk potetial.

The general rule and the fundamental difference for disk potentials and their rotation curves is that the mass contained at radii larger than R has a large effect on the circular velocity at R.

```python
from galpy import potential
class EquivalentMassSphericalPotential(potential.Potential):
    def __init__(self,amp=1.,cumulmassfunc=None,ro=None,vo=None):
        potential.Potential.__init__(self,amp=amp,ro=ro,vo=vo)
        self._cumulmassfunc= cumulmassfunc
    def _Rforce(self,R,z,phi=0.,t=0.):
        r= numpy.sqrt(R**2.+z**2.)
        return -self._cumulmassfunc(r)*R/r**3.

from galpy import potential
from galpy.util import bovy_plot
figsize(10,6)
a= 1.
kzp= potential.KuzminDiskPotential(amp=5.,a=a)
kzp_spher= EquivalentMassSphericalPotential(amp=5.,\
                cumulmassfunc=lambda R: R/numpy.sqrt(R**2+a**2))
line_kuzmin= kzp.plotRotcurve(Rrange=[0.,15.],\
                label=r'$\mathrm{Kuzmin}$',yrange=[0.,2.5],
                xlabel=r'$R/a$',ylabel=r'$v_c(R/a)$')
line_kuzmin_spher= kzp_spher.plotRotcurve(Rrange=[0.,15.],\
                label=r'$\mathrm{Spherical\ with\ same}\ M(<R)$',
                overplot=True)
legend(handles=[line_kuzmin[0],line_kuzmin_spher[0]],
       fontsize=18.,loc='upper right',frameon=False);
```

## Leapfrog

I decided to use the leapfrog method of integration. The basic idea of the leapfrog method is to stagger the updates of positions and velocity in time. Doing this makes the method more symmetric and therefore more accurate. 

In a spherically symmetric potential the total angular momentum is conserved. Here, I first look at a 2D disk which should also conserve angular momentum. One of the advantages of the leapfrog method is that it conserves angular momentum exactly although it does not conserve energy exactly.


### 2D disk

We can take a look at this now by running the simulation below,




```python
from IPython import display
import time
import numpy as np
import matplotlib.pyplot as plt

G=1.
N=500 #number of test particles
m=1. #mass test particle
M=N*m #Total mass of system
r=np.random.random(N)
theta=2*np.pi*np.random.random(N)
Lang=np.array([])
x=np.zeros(N)
y=np.zeros(N)
for i in range(N):
    x[i]=r[i]*np.cos(theta[i])
    y[i]=r[i]*np.sin(theta[i])
vx=np.zeros(N)
vy=np.zeros(N)

t=0
tmax=100
dt=0.001

plt.ion()
fig, ax = plt.subplots()

while t<tmax:
    #print (t)
    for i in range(N):
        x[i] += vx[i]*dt/2.
        y[i] += vy[i]*dt/2.
    for i in range(N):
        ax=0.
        ay=0.
        for j in range(N):
            if i!=j:
                dx=x[i]-x[j]
                dy=y[i]-y[j]
                r=np.sqrt(dx*dx+dy*dy)
                a=-G/(r*r)
                ax+=a/r*dx
                ay+=a/r*dy
        vx[i]+=ax*dt
        vy[i]+=ay*dt
        L=0
    for i in range(N):
        x[i]+=vx[i]*dt/2
        y[i]+=vy[i]*dt/2
        L+=x[i]*vy[i]-y[i]*vx[i] # calculating the total angular momentum (L = r X mv) at each time step.
    #print (L)
    Lang=np.append(Lang,L)
    t=t+dt
    plt.scatter(x,y)
    plt.xlim([-2,2])
    plt.ylim([-2,2])
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(0.05)
    plt.clf()
#     plt.pause(0.1)
#     plt.draw()
#     plt.clf()
```

```python
from IPython import display
import time
import numpy as np
import matplotlib.pyplot as plt

G=1.
N=500 #number of test particles
m=1. #mass test particle
M=N*m #Total mass of system
r=100*np.random.random(N)
theta=2*np.pi*np.random.random(N)
Lang=np.array([])
Mcore=10000 #massive core
x=np.zeros(N)
y=np.zeros(N)
for i in range(N):
    x[i]=r[i]*np.cos(theta[i])
    y[i]=r[i]*np.sin(theta[i])
vx=np.zeros(N)
vy=np.zeros(N)

for i in range(N): # This is where I calculate initial velocities
    rcore=np.sqrt(x[i]*x[i]+y[i]*y[i])
    acore=G*Mcore/(rcore*rcore)
    v=np.sqrt(acore*rcore)
    vx[i]=-v/rcore*y[i]
    vy[i]=v/rcore*x[i]
tdyn=np.median(2*np.pi*np.sqrt(x**2+y**2)/(vx**2+vy**2)) #rough estimate of the dynamical time
t=0
tmax=100*tdyn
dt=tdyn/50 #50 time-steps within an orbit

#plt.ion()
fig, ax = plt.subplots()
step=0
while t<tmax:
    data=np.column_stack((x,y))
    if (t==0):
        plt.scatter(x,y)
        plt.xlim([-150,150])
        plt.ylim([-150,150])
        #plt.pause(0.1)
        #plt.draw()
        #np.savetxt('plots3/time0.0.dat', data)
        #plt.savefig('plots3/test0.0.png')
        plt.clf()
        step+=1
        t=t+dt
    else:
        #print (t)
        for i in range(N):
            x[i] += vx[i]*dt/2.
            y[i] += vy[i]*dt/2.
        for i in range(N):
            ax=0.
            ay=0.
            for j in range(N):
                if i!=j:
                    dx=x[i]-x[j]
                    dy=y[i]-y[j]
                    r=np.sqrt(dx*dx+dy*dy)
                    a=-G/(r*r)
                    ax+=a/r*dx
                    ay+=a/r*dy
            rcore=np.sqrt(x[i]*x[i]+y[i]*y[i])
            axcore=-G*Mcore*x[i]/(rcore*rcore*rcore)
            aycore=-G*Mcore*y[i]/(rcore*rcore*rcore)
            vx[i]+=(ax+axcore)*dt
            vy[i]+=(ay+aycore)*dt
        L=0
        for i in range(N):
            x[i]+=vx[i]*dt/2
            y[i]+=vy[i]*dt/2
            L+=x[i]*vy[i]-y[i]*vx[i] # calculating the total angular momentum (L = r X mv) at each time step.
        Lang=np.append(Lang,L)
        #if (np.remainder(step,10)==0):
        plt.scatter(x,y)
        plt.xlim([-150,150])
        plt.ylim([-150,150])
        plt.title('t=%.1f $t_{dyn}$' %(step/50))
        #plt.pause(0.1)
        #plt.draw()
        #np.savetxt('plots3/time%.1f.dat' %(step/50), data)
        #plt.savefig('plots3/test%.1f.png' %(step/50))
        display.clear_output(wait=True)
        display.display(plt.gcf())
        time.sleep(0.05)
        plt.clf()
            
            
        step+=1
        t=t+dt
```
